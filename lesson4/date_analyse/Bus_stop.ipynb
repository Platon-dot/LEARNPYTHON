```python
import pandas as pd
stations = pd.read_csv(
    open('bus_stops.csv', 'r', encoding='cp1251'),
    sep=';'
)
```

```python
stations.head(10)
```

```python
stations_count = stations.groupby(['Street']).size()
```

```python
stations_count
```

```python
type(stations_count)
```

```python
stations_count.sort_values(ascending=False, inplace=True)  #сортировка полученных значений , inplace - с записью к конкретную переменную
```
