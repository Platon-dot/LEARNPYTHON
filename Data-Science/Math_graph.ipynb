```python
import random
import matplotlib.pyplot as plt #сокращение название команды
from matplotlib.cm import get_cmap
import pandas as pd
```

```python
data_x1 = range(1, 100, 5)
data_y1 = range(0, 20)
# вклбчение графика
plt.plot(data_x1, data_y1)
plt.show()
```

```python
# использование молудя sample
data_x2 = random.sample(range(500), 10)
data_y2 = random.sample(range(100), 10)
data_x2.sort()
data_y2.sort()
plt.plot(data_x2, data_y2)
plt.show()
```

```python
plt.plot(data_x1, data_y1, color='red', linestyle='dashed')
# если я хочу отрисовать рядом два графика а не накладывать один на другой
plt.figure()
plt.plot(data_x2, data_y2, color='green', linestyle='dotted', linewidth=5)
plt.show()
```

```python
stations = pd.read_csv(open('bus_stops.csv', 'r', encoding='cp1251'), sep=';')
```

```python
stations 
```

```python
pavilion_count = stations.groupby(['Pavilion']).size()
```

```python
pavilion_count
```

```python
plt.pie(pavilion_count, labels=pavilion_count.index, autopct='%.1f')
plt.title('Павильоны у остановок')
plt.legend()
plt.show()
```

```python
admarea_count = stations.groupby(['AdmArea']).size()
```

```python
admarea_count
```

```python
labels = ['Noname',
          'ВАО',
          'ЗАО',
          'Зел',
          'Нов',
          'САО',
          'СВАО',
          'СЗАО',
          'Тро',
          'ЦАО',
          'ЮВАО',
          'ЮЗАО',
          'ЮО'
         ]
# передаем название цветовой схемы
cmap = get_cmap('tab20')
# так как цвета нельзя взять напрямую, то берем цвета для каждого столбика
colors = [cmap(i) for i in range(len(lables))]
plt.bar(labels, admarea_count, color=colors)
plt.title('Остановки по округам')
plt.show
```
